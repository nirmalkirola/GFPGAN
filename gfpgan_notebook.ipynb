```python
import matplotlib.pyplot as plt
from gfpgan_modified import main
import cv2
```

```python
img_path = 'test_img/1.jpg.webp'
```

```python
# main function from gfpgan_modified

original_face, restored_face, restored_image = main(img_path)
```

```python
og_image = cv2.imread(img_path)
og_image = cv2.cvtColor(og_image, cv2.COLOR_BGR2RGB)
```

```python

plt.figure()
plt.title('original face')
plt.imshow(original_face)

plt.figure()
plt.title('restored face')
plt.imshow(restored_face)

plt.figure()
plt.title('original image')
plt.imshow(og_image)

plt.figure()
plt.title('restored image')
plt.imshow(restored_image)
```
